### Data Visualization
Moving on to visualizing the trends and relations.  

#### **Relationship between Steps and Sedentary Time**  

````r
#merging data-sets
  merge(sleep, dailyActivity, by = "Id", all = "TRUE")
#Visualization-1(relationship b/w steps and sedentary time)
  ggplot(data = dailyActivity, aes(x=TotalSteps, y = SedentaryMinutes)) + geom_point() + geom_smooth() + labs(title = "Total Steps vs Sedentary Minutes")
  cor(dailyActivity$TotalSteps, dailyActivity$SedentaryMinutes)
````

As we can see from the graph and the stat, there is a negative correlation between Steps taken in a dat and Sedentary Time, which implies that more steps you take, the less sedentary lifestyle you lead. This data shows that the company need to market more the customer segments with **high sedentary time** and to do that the company needs to find ways to get started in walking more and also measure their daily steps.  


#### **Relationship between Minutes Asleep and Time in Bed**  

````r
#visualization-2(relationship b/w minutes asleep and time in bed)
  ggplot(data = sleep, aes(x=TotalMinutesAsleep, y = TotalTimeInBed)) + geom_point() + geom_smooth() + labs(title = "Minutes Asleep vs Time in Bed(in minutes)")
  cor(sleep$TotalMinutesAsleep, sleep$TotalTimeInBed)
````

As we can see from the graph and the stat, there is a positive correlation between Minutes Asleep and Time in Bed. So, a marketing strategy that I would propose would be to help users sleep by sending out notifications on their device which would ask them to **go to sleep**.  


#### **Relationship between Calories burnt and time**

````r
#merging data-sets
  Combined_data = merge(weight, dailyCalories, by = "Id", all = "TRUE")

#visualization-3(relationship b/w calories burnt vs time)
  Combined_data$Date = format(Combined_data$Date, format = "%H:%M:%S")
  ggplot(data = Combined_data, aes(x=Date, y = Calories)) + geom_histogram(stat = "identity", fill = "darkblue") + theme(axis.text.x = element_text(angle = 90)) + labs(title = "Calories burnt vs Time")
````

By analyzing the dailyCalories and weight dataset overtime, the company will have a good idea on how customers are using their product during the day. Most users are active before and after the work(according to the data provided). The company can use the **Bellabeat app** to remind and motivate users to *log a activity* during that time more.
